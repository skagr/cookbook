````python
import pandas as pd
import seaborn as sns
````

````python
df = sns.load_dataset('titanic')
df.head()
````

### Simple selections

Selecting rows based on a single column is easy:

````python
df.loc[df.survived == 1].head()
````

### Boolean logic

Let's say we want to combine 2 filters. Python AND/OR operators don't work on Pandas Series, and the error message is confusing (at least it was to me at first):

```
>>> df.loc[(df.survived == 1) and (df.sex == 'female')]
...
ValueError: The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all().
```

We need to use [bitwise operators](https://wiki.python.org/moin/BitwiseOperators) like **`&`** and **`|`** here:

````python
# Female survivors
df.loc[(df.survived == 1) & (df.sex == 'female')].head()
````

````python
# Women and children under 10
df.loc[(df.sex == 'female') | (df.age < 10 )].head()
````

### Complex selections

Pandas' `.loc` method filters on booleans. Knowing that, we can use any function that returns a Series of boolean values with len(df).

````python
def single_male_survivor_outliers(df, col):
    sd = df[col].std()
    mean = df[col].mean()
    outlier = (df[col] < mean - 2 * sd) | (df[col] > mean + 2 * sd)
    return df.alone & (df.sex == 'male') & (df.survived == 1) & outlier

df.loc[single_male_survivor_outliers(df, 'age')].head()
````

### A note on looping

The function above takes in a Dataframe and returns a Series of True/False values. Every operation in this function works on an entire Series, so there's no looping. This type of operation is called 'vectorized', and is generally faster than row-wise operations.

You can still loop over all rows using `.apply(axis=1)` `.iterrows()` and `.itertuples()`, but this is very rarely necessary, and almost certainly slower.
