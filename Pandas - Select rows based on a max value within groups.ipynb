```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

If we just wanted the max, age per group we could do this:

```python
df.groupby("embark_town")["age"].max()
```

But what if we want the original rows based on the maximum age per group?

#### Select rows based on the *first occurence* of max value of age per embark town

```python
df.loc[df.groupby("embark_town")["age"].idxmax()]
```

#### Select rows based on max values of age per embark town, with duplicates

```python
df[df.groupby(['embark_town'])['age'].transform(max) == df['age']]
```
